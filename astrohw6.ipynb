```python
h = 6.626E-34
c = 3E8
k = 1.38E-23

def planckl(lam, T):
    Bl1 = 2*h*c**2
    Bl2 = h*c/(lam*k*T)
    B = Bl1/((lam**5)*(np.exp(Bl2)-1))
    return B

def planckh(Hz, T):
    Bh1 = 2*h*Hz**3
    Bh2 = h*c/(lam*k*T)
    B = Bh1/((lam**5)*(np.exp(Bh2)-1))
    return B
```

```python
lam1 = np.arange(1e-9,8e-6,1e-9) 
lam = np.arange(1e-9,4e-6,1e-9) 

dat1= planckl(lam1, 10)
plt.plot(lam1*1e9, dat1)

plt.yscale("log")
#plt.xscale("log")
```
